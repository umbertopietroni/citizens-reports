```python
amb = segnalazioni.loc[segnalazioni['infoclass'] == "Ambiente"][:1400]
sic = segnalazioni.loc[segnalazioni['infoclass'] == "Sicurezza"][:1400]
ill = segnalazioni.loc[segnalazioni['infoclass'] == "Illuminazione"][:1400]
man = segnalazioni.loc[segnalazioni['infoclass'] == "Manutenzione"][:1400]


result = amb.append([sic,ill,man])
```

```python
result
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test = train_test_split(result,test_size=0.3, random_state=1)
```
